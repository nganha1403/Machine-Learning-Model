```python
import pandas as pd
```

```python
df = pd.read_csv("Data WorldBank fix.csv")
print(df)
```

```python
df = df.drop(['Inflation', 'Exchange Rate','Money Supply', 'Unemployment', 'FDI'], axis=1)
print(df)
```

```python
filtered_df = df[df['Country Name'].str.contains('Viet Nam', na=False)]
```

```python
print(filtered_df)
```

```python
filtered_df['GDP'] = filtered_df['GDP'].apply(lambda x: f"{x:.0f}")
print(filtered_df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 20))
filtered_df.plot()
plt.show()
```

```python
filtered_df['GDP'] = filtered_df['GDP'].astype('float64')
filtered_df['GDP']
```

```python
gdp_sample_df = filtered_df.drop(columns=['Country Name'])
gdp_sample_df
```

```python
gdp_sample_df['Year'] = gdp_sample_df['Year'].apply(lambda x: str(x).split('-')[0])
gdp_sample_df['Year']
```

```python
gdp_sample_df['Year'] = pd.to_datetime(gdp_sample_df['Year'])
gdp_sample_df['Year']
```

```python
gdp_sample_df1 = gdp_sample_df.set_index('Year')
gdp_sample_df1
```

```python
gdp_sample_df1.info()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test2(x):
    result = adfuller(x)
    labels = ['ADF Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
    if result[1] <=0.05:
        print('strong evidence against null hypothesis, we will reject null hypothesis indicating it is stationary')
    else:
        print('due to weak evidence or lack of evidence against null hypothesis, we fail to reject null hypothesis indicating it is non-staionary')
```

```python
adfuller_test2(gdp_sample_df1['GDP'])
```

```python
from statsmodels.tsa.arima_model import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

```python
print(gdp_sample_df1.shape)
train = gdp_sample_df1.iloc[:-3]
test = gdp_sample_df1.iloc[-3:]
print(train.shape, test.shape)
print(len(train))
```

```python
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')


train_model = ARIMA(train['GDP'], order=(1,2,1))
train_model = train_model.fit()
```

```python
## and now predict on test set

start_train = len(train)
end_train = len(train)+len(test)-1
p = train_model.predict(start=start_train, end=end_train, typ='levels')
print(p)
```

```python
p.plot(legend=True)
test['GDP'].plot(legend=True)
```

```python
test['GDP'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(p,test['GDP']))
print(rmse)
```

```python
model_df = ARIMA(gdp_sample_df1['GDP'], order=(1,2,1)) #11,2,11 is also valid
result = model_df.fit()
result.summary()
#gdp_sample_df1.tail()
```

```python
from statsmodels.graphics.tsaplots import plot_predict # import the plot_predict function

plot_predict(result, start=14, end=20, dynamic=False)
```
